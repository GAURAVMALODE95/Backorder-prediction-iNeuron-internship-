```python
'''A backorder is an order for a good or service that cannot be filled at the current time due to a lack of available supply. The item may not be held in the company's available inventory but could still be in production, or the company may need to still manufacture more of the product.
The backorder is an indication that demand for a company's product outweighs its supply. They may also be known as the company's backlog. Also, company cannot overstock every product in their inventory to avoid such situation.'''
```

sku: It is a random ID assigned to each product. This column is used for identification purposes.

national_inv: This column represents the current inventory level of the product, indicating how many units are available in stock.

lead_time: It refers to the transit time for the product, indicating the time taken for the product to be delivered if available.

in_transit_qty: This column represents the quantity of the product that is currently in transit from the source to the destination.

forecast_3_month: It indicates the forecasted sales quantity for the next 3 months for the specific product.

forecast_6_month: Similar to the previous column, it represents the forecasted sales quantity for the next 6 months.

forecast_9_month: This column indicates the forecasted sales quantity for the next 9 months.

sales_1_month: It represents the actual sales quantity for the prior 1 month time period.

sales_3_month: This column indicates the actual sales quantity for the prior 3 month time period.

sales_6_month: Similar to the previous column, it represents the actual sales quantity for the prior 6 month time period.

sales_9_month: This column indicates the actual sales quantity for the prior 9 month time period.

min_bank: It represents the minimum recommended amount of stock for the specific product.

potential_issue: This column identifies the source issue for the part, indicating whether there are any potential issues associated with it.

pieces_past_due: It represents the number of parts that are overdue from the source.

perf_6_month_avg: This column indicates the source performance for the prior 6-month period.

perf_12_month_avg: Similar to the previous column, it represents the source performance for the prior 12-month period.

local_bo_qty: It indicates the amount of stock orders that are currently overdue.

deck_risk: This column represents a part risk flag, indicating whether the specific part is at risk.

oe_constraint: Similar to the previous column, it represents a part risk flag, indicating whether there are any constraints associated with the part.

ppap_risk: This column represents a part risk flag, indicating whether there is a risk associated with the part.

stop_auto_buy: Similar to the previous column, it represents a part risk flag, indicating whether there is a need to stop auto-buy for the part.

rev_stop: This column represents a part risk flag, indicating whether there is a revenue stop associated with the part.

went_on_backorder: This is the target column you mentioned. It indicates whether the product actually went on backorder or not. It is the value you want to predict using the other columns as features.

```python
rows_with_null = df[df.isnull().any(axis=1)]

# Get the row indices with null values
row_indices = rows_with_null.index

# Remove the rows with null values
df_cleaned = df.drop(row_indices, axis=0)

# Verify the shape of the cleaned dataset
print(df_cleaned.shape)
```

```python
df_cleaned.head()
```

```python
df_cleaned.isnull().sum()  # data is fully cleaned
```

```python
categorical_cols = ["potential_issue", "deck_risk", "oe_constraint", "ppap_risk", "stop_auto_buy", "rev_stop", "sku"]
mapping_dict = {
    "potential_issue": {"No": 0, "Yes": 1},
    "deck_risk": {"No": 0, "Yes": 1},
    "oe_constraint": {"No": 0, "Yes": 1},
    "ppap_risk": {"No": 0, "Yes": 1},
    "stop_auto_buy": {"No": 0, "Yes": 1},
    "rev_stop": {"No": 0, "Yes": 1},
    "went_on_backorder": {"No": 0, "Yes": 1}
}
df_cleaned.replace(mapping_dict, inplace=True)
```
